```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (10, 10))
sns.histplot(x=' Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i,z in enumerate([' Estimated Unemployment Rate (%)',' Estimated Employed', ' Estimated Labour Participation Rate (%)']):
    plt.subplot(2,2,i+1)
    sns.histplot(x=z, data=df, kde=True,hue="Area")
    
    plt.ylabel("Count")
plt.title("Histogram of Estimated Employment Rate by Area")
plt.show()
```

```python
sns.pairplot(df,hue =  "Area")
```

```python
plt.figure(figsize =  (15,6))
data_grouped = df.groupby('Region')[' Estimated Unemployment Rate (%)'].sum()
data_grouped.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Estimated Unemployment Rate by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
sns.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (9, 9))
plt.scatter(df2['Region.1'], df2[' Estimated Employed'])
plt.title('Employment according to Region')
plt.xlabel('Region')
plt.ylabel('Employment Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
df2.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=df2)
plt.show()
```
